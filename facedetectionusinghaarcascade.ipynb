```python
import cv2
```

```python
image=cv2.imread('/content/people1.jpg')
```

```python
image.shape
```

```python
image=cv2.resize(image,(800,600));
image.shape
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```python
image_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY);
cv2_imshow(image_gray)
```

```python
face_detector=cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
```

```python
detections=face_detector.detectMultiScale(image_gray,scaleFactor=1.09)
```

```python
detections
```

```python
len(detections)
```

```python
for(x,y,w,h) in detections:
#  cv2.rectangle(image,(x,y),(x+w,y),(0,0,255),10);
 cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,255),10);
```
